```python
# Removing the missing values
newdf=df[['Review Text','Recommended IND']]
newdf.dropna(inplace=True)
print('Dimension of the new data set')
newdf.shape
```

```python
# Determining the sentiment analysis of the data
# Considering the data of only first 18000 rows
sentiments=newdf['Recommended IND'].values[:18000]
reviews=newdf['Review Text'].values[:18000]
```
